```python
def totalExp(listOfExp):
    totalYears = 0
    totalMonths = 0
    for exp in listOfExp:
        if exp != '':
            if 'Expire' not in exp:
                try:
                    if exp.split()[1] == 'mos' or exp.split()[1] == 'mo':
                        totalMonths = totalMonths + int(exp.split()[0])
                    elif exp.split()[3] == 'mos' or exp.split()[3] == 'mo':
                        totalMonths = totalMonths + int(exp.split()[2])
                        totalYears = totalYears + int(exp.split()[0])
                except:
                    totalYears = totalYears + int(exp.split()[0])
    
    totalMonths = totalMonths%12
    totalYears  = totalYears + (totalMonths//12)   
    
    return (str(totalYears) + ' years, ' + str(totalMonths) + ' months')
```

```python
df['Years of Experience'] = df['Years of Experience'].map(lambda x: totalExp(x.replace('[','').replace(']','').replace("'",'').split(',')))
```

```python
df['Top Skills'] = df['Top Skills'].map(lambda x: x.replace(']','')
                                        .replace('[','')
                                        .replace("'",''))
```

```python
df['About'] = df['About'].map(lambda x: x if str(x) == 'nan'
                              else x.replace('\n','')
                              .replace('see more', '')
                              .replace('…', '')
                              .strip())
```

```python
df['Location'] = df['Location'].map(lambda x: x.replace('\n','').strip())
```

```python
df['Index'] = df.reset_index()['index'].map(lambda x: 'C' + str(x+1))

df = df[['Index', 'Location', 'Top Skills', 'Years of Experience']]

df['Location'] = df['Location'].map(lambda x: x.lower())


def cleanLocation(location):

    locationList = location.split(', ')
    if locationList[0] != 'india':
        return ", ".join([name for name in locationList if name != 'india'])
    else:
        return ", ".join([name for name in locationList])


df['Location'] = df['Location'].map(lambda x: cleanLocation(x))

df = cc.cityToZoneMapping(df)


def experienceCalculator(exp):
    year, month = exp.split(", ")[0].split(
        " ")[0], exp.split(", ")[1].split(" ")[0]

    totalExp = (int(year)*12 + int(month))/12

    return round(totalExp)


df['Years of Experience'] = df['Years of Experience'].map(
    lambda x: experienceCalculator(x))

df.rename(columns={'Years of Experience': 'Experience'}, inplace=True)


# print(df)
df.to_csv('./data/finalTestData.tsv', sep='\t', index=False)
```
