```python
import pandas as pd
import matplotlib.pyplot as plt
import utils.cityCleaning
import utils.MaxSalary

# data from job portal, individual results of queries ran on different cities are grouped together

df = pd.read_excel("./data/Final_Dataset.xlsx", engine='openpyxl')

# treating NaN values and combining logical columns

df['Company'] = df[['Company', 'company']].apply(lambda x: 
                                 x['Company'] if str(x['Company']) != 'nan' else x['company'], 
                                 axis=1)

# drop redundant columns

df.drop(['Se.No','company', 'Description'], axis=1, inplace=True)
```

```python
# exrtract max salary from a function utility created
df['Salary'] = df['Salary'].map(lambda x: utils.MaxSalary.maxSalary(x))
```

```python
skillSetDict = dict()
```

```python
# remove extra characters from the list of skills which is present as string
def createSkillSet(skillstring):
    for skill in skillstring.replace('[','').replace(']','').replace("'","").split(','):
        skillSetDict[skill.strip().lower()] = 1 + skillSetDict.get(skill.strip().lower(), 0)
```

```python
df['Skillset'].map(lambda x: createSkillSet(x))
```
